<h3>First observation is to find the average Delay of a particular airline. From the above infomration, I can see there are total of 31 columns in flights_details but only two of them seems relevant to my question (i.e. ARRIVAL_DELAY, DEPARTURE_DELAY, AIRLINE_DELAY). Now, AIRLINE_DELAY has a huge number of missing values so I'd rather not want to use it. Now, I'd consider AVERAGE_DELAY=ARRIVAL_DELAY+DEPARTURE_DELAY (for a particular airline)


```python
average_arrival_delay = pd.DataFrame({'average_arrival_delay': flights_details.groupby(['AIRLINE'])['ARRIVAL_DELAY'].mean()})
average_arrival_delay
```

```python
average_departure_delay = pd.DataFrame({'average_departure_delay': flights_details.groupby(['AIRLINE'])['DEPARTURE_DELAY'].mean()})
average_departure_delay
```

```python
monthwise_flight_arrival_delay = pd.DataFrame({'Delay Monthwise': flights_details.groupby(['MONTH'])['ARRIVAL_DELAY'].mean()})
monthwise_flight_arrival_delay
```

```python
monthwise_flight_departure_delay = pd.DataFrame({'Delay Monthwise': flights_details.groupby(['MONTH'])['DEPARTURE_DELAY'].mean()})
monthwise_flight_departure_delay
```

```python
monthly_delay = pd.concat([monthwise_flight_departure_delay, monthwise_flight_arrival_delay])
monthly_average_delay = monthly_delay.groupby(level=0).mean()
monthly_average_delay
```

```python
monthwise_flights_cancelled = flights_details[flights_details['CANCELLED'] == 1].groupby('MONTH').count()['CANCELLED']
monthwise_flights_cancelled
```

<h3>Above observations show that there are a lot of number of cancellations in the first three months and last month of the year has a lot of cancellations in flights as compared to other months. <h4>Now the ranking approach for airports.

To rank the airport, first of all the weather in that city (location where airport is located) is quite important. Major reason for flight cancellations/diversions/delays are generally due to weather issues. The Airport located in a city where weather is favorable in most of the months would be considered good. The airport with major number of terminals/gates suggests that it is a large/small airport where more number of gates suggests that it is connected internationally to major countries. Now, the ratio of mean_of_flight_delays(arival and departure)/total_number_of_flights(arrival and departure) can help to give us the idea of average_delay for a flight in that particular airport. Lesser the value of ratio, more the chances of airport to be ranked on top. The air_traffic can also play one important factor. More the air_traffic value suggests the busier the airport would be. On and all, I would consider the following features to come up with the ranking of airport.

DELAY_IN_FLIGHTS (ARRIVAL_DELAY + DEPARTURE_DELAY)
TOTAL_FLIGHTS (ARRIVAL + DEPARTURE)
AIR_TRAFFIC_VALUE (DAILY/WEEKLY/YEARLY)
TEMPERATURE_ON_VARIOUS_MONTHS (WHERE AIRPORT IS LOCATED)
NUMBER_OF_GATES/TERMINALS_OF_AIRPORT
NUMBER_OF_CANCELLED/DELAYED_FLIGHTS_FROM_AIRPORT
